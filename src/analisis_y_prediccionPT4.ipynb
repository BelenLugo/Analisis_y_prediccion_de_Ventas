```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../data/retail_sales_dataset.csv')
```

```python
#Observamos las caracteristicas
df.describe()
```

```python
# Histograma y boxplot para 'Total Amount'
plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
sns.histplot(df['Total Amount'], kde=True, bins=30, color='blue')
plt.title('Distribuci√≥n de Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frecuencia')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Total Amount'], color='orange')
plt.title('Boxplot de Total Amount')

plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por fecha
ventas_por_fecha = df.groupby('Date')['Total Amount'].sum()

# Gr√°fico de l√≠neas
plt.figure(figsize=(12, 6))
ventas_por_fecha.plot(kind='line', marker='o', color='green')
plt.title('Tendencia de Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Total de Ventas')
plt.grid()
plt.show()
```

```python
# Gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price per Unit', y='Total Amount', data=df, hue='Product Category', palette='viridis')
plt.title('Relaci√≥n entre Price per Unit y Total Amount')
plt.xlabel('Price per Unit')
plt.ylabel('Total Amount')
plt.legend(title='Product Category')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Histograma
plt.subplot(2, 1, 1)
sns.histplot(df['Total Amount'], kde=True, bins=30, color='blue')
plt.title('Histograma de Total Amount')

# Boxplot
plt.subplot(2, 1, 2)
sns.boxplot(x=df['Total Amount'], color='red')
plt.title('Boxplot de Total Amount')

plt.tight_layout()
plt.show()
```
